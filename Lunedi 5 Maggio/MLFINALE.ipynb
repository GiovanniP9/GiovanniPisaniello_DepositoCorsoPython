## Capitolo 1 – Introduzione al Machine Learning

### Cos'è il Machine Learning

Il Machine Learning (ML), o apprendimento automatico, è un campo dell’informatica che consente ai computer di apprendere da dati ed esperienza, senza essere esplicitamente programmati. In altre parole, un sistema di Machine Learning è in grado di riconoscere schemi nei dati e prendere decisioni o fare previsioni basandosi su di essi.

A differenza dei metodi tradizionali in cui il programmatore specifica passo passo cosa deve fare il computer, con il Machine Learning si forniscono al sistema grandi quantità di dati e un algoritmo capace di apprendere da questi dati per costruire un modello predittivo.

Un classico esempio è la casella di posta elettronica che impara a distinguere i messaggi di spam da quelli legittimi. Inizialmente, l’algoritmo riceve esempi etichettati come “spam” e “non spam”. Confrontando le caratteristiche dei messaggi, l’algoritmo impara a classificare automaticamente le nuove email.

### Perché si usa il Machine Learning

Il Machine Learning si usa perché consente di risolvere problemi complessi che sarebbero molto difficili o impossibili da risolvere con la programmazione tradizionale. È particolarmente utile quando:

- I dati sono abbondanti e complessi
- Le regole per risolvere un problema non sono note o sono troppo numerose per essere codificate manualmente
- Si vuole ottenere un sistema capace di adattarsi nel tempo e migliorare con l’esperienza

I modelli di Machine Learning sono anche capaci di generalizzare, ovvero di applicare ciò che hanno appreso su un insieme di dati a nuovi dati mai visti prima. Questa capacità è fondamentale per applicazioni reali dove i dati cambiano continuamente.

### In cosa è utile il Machine Learning

Il Machine Learning trova applicazione in moltissimi settori. Ecco alcuni esempi pratici:

- **Visione artificiale:** riconoscimento facciale, rilevamento di oggetti, classificazione di immagini
- **Elaborazione del linguaggio naturale:** traduzione automatica, analisi del sentiment, chatbot
- **Finanza:** rilevamento di frodi, valutazione del rischio, previsione dei mercati
- **Medicina:** diagnosi automatica da immagini mediche, analisi di dati clinici, previsione di epidemie
- **Marketing:** personalizzazione di contenuti, raccomandazioni di prodotti, analisi dei clienti
- **Automobili autonome:** rilevamento dell’ambiente circostante, pianificazione del percorso, decisioni in tempo reale

In sintesi, il Machine Learning è uno strumento potente per estrarre valore dai dati e costruire sistemi intelligenti, capaci di apprendere e migliorare nel tempo.


## Capitolo 2 – Tipi di Apprendimento Automatico

Il Machine Learning si suddivide principalmente in tre categorie fondamentali, ognuna delle quali si basa sul tipo di dati forniti al sistema e sul tipo di problema da risolvere:

### 1. Apprendimento Supervisionato (Supervised Learning)

Nell’apprendimento supervisionato, al modello viene fornito un insieme di dati etichettati, ovvero ogni esempio è accompagnato dalla risposta corretta. Lo scopo è addestrare un modello che sappia prevedere l’etichetta corretta per nuovi dati mai visti.

#### Esempio
Si vuole costruire un sistema che preveda il prezzo di una casa. Ogni esempio nel dataset contiene caratteristiche come la dimensione, il numero di stanze, la posizione e il prezzo effettivo della casa. Il modello impara a stimare il prezzo in base a queste caratteristiche.

#### Applicazioni
- Previsione dei prezzi (case, azioni, prodotti)
- Riconoscimento di immagini etichettate (es. cane, gatto)
- Classificazione di email come “spam” o “non spam”

#### Tipi di output
- **Regressione:** quando l’output è un numero continuo (es. prezzo, temperatura)
- **Classificazione:** quando l’output è una categoria (es. malato/sano)

---

### 2. Apprendimento Non Supervisionato (Unsupervised Learning)

Nell’apprendimento non supervisionato, i dati forniti al modello non sono etichettati. L’obiettivo è scoprire strutture nascoste, raggruppamenti o schemi all’interno dei dati.

#### Esempio
Si dispone di dati relativi ai comportamenti di acquisto di clienti, ma senza etichette. Il sistema può individuare gruppi (cluster) di clienti con comportamenti simili per proporre offerte personalizzate.

#### Applicazioni
- Segmentazione di clienti
- Compressione dei dati
- Rilevamento di anomalie (es. frodi)
- Analisi esplorativa

#### Algoritmi comuni
- K-means
- PCA (Principal Component Analysis)
- DBSCAN

---

### 3. Apprendimento per Rinforzo (Reinforcement Learning)

Nell’apprendimento per rinforzo, un agente apprende interagendo con un ambiente. L’agente compie delle azioni e riceve ricompense o penalità in base all’esito. L’obiettivo è massimizzare la ricompensa totale nel tempo.

#### Esempio
Un robot che deve imparare a camminare riceve una ricompensa quando fa un passo in avanti e una penalità quando cade. Attraverso tentativi ed errori, il robot impara la sequenza ottimale di azioni.

#### Applicazioni
- Giochi (scacchi, Go, videogiochi)
- Robotica
- Finanza algoritmica
- Sistemi di raccomandazione dinamici

---

### Confronto tra i tipi di apprendimento

| Tipo               | Input                      | Obiettivo                            | Esempi                                   |
|--------------------|----------------------------|--------------------------------------|------------------------------------------|
| Supervisionato     | Dati con etichette         | Prevedere una risposta               | Classificare email, prevedere prezzi      |
| Non supervisionato | Dati senza etichette       | Trovare strutture o gruppi           | Segmentazione clienti, rilevamento anomalie |
| Per rinforzo       | Interazione con l’ambiente | Massimizzare ricompense nel tempo    | Robotica, giochi, guida autonoma          |

## Capitolo 3 – L’importanza dei Dati: Quando la Semplicità Vince

Uno dei principi fondamentali del Machine Learning moderno è che **la quantità e la qualità dei dati hanno un impatto maggiore delle complessità algoritmiche**, specialmente quando i dataset sono molto ampi. Questo concetto è stato dimostrato sia teoricamente che empiricamente in numerosi studi, e ha influenzato profondamente il modo in cui si progettano i sistemi di apprendimento automatico oggi.

### Algoritmi semplici e grandi quantità di dati

Negli anni 2000, una serie di ricerche ha mostrato che **algoritmi semplici, come il k-Nearest Neighbors, le regressioni lineari o i modelli a bag-of-words**, potevano raggiungere o addirittura superare le performance di modelli più sofisticati **quando venivano alimentati con grandi quantità di dati di alta qualità**.

#### Esempio pratico
Google ha dimostrato in diverse occasioni che **un semplice modello n-gramma** (un metodo statistico per il linguaggio naturale) poteva battere modelli linguistici molto più complessi, semplicemente perché era addestrato su un corpus molto più grande di testo. Allo stesso modo, i sistemi di traduzione automatica iniziali di Google si basavano su modelli relativamente semplici che beneficiavano enormemente di dati multilingua su scala planetaria.

### Perché succede

Il motivo per cui gli algoritmi semplici possono performare così bene su grandi dataset risiede in alcuni concetti chiave:

- **Generalizzazione naturale:** algoritmi semplici hanno meno parametri da ottimizzare e sono meno soggetti a overfitting. Quando i dati sono molti, questo vantaggio si amplifica.
- **Rumore ridotto dal volume:** grandi dataset permettono di compensare la presenza di rumore o errori nei dati, facendo emergere la struttura statistica sottostante.
- **Maggiore copertura dei casi rari:** i dati abbondanti coprono meglio le eccezioni e i casi particolari, rendendo meno necessario l’uso di strutture complesse per coglierli.

### Legge dell’asintotica nei dati

Uno dei risultati più noti in questo ambito è la cosiddetta “**legge dell’asintotica dei dati**”:  
> *"Per molti problemi pratici, un modello semplice ben addestrato su molti dati performa meglio di un modello complesso addestrato su pochi dati."*

Questo principio è alla base del successo di molte aziende nel settore tecnologico. Non a caso, aziende come Google, Facebook, Amazon e OpenAI investono enormi risorse nella raccolta, pulizia e gestione dei dati.

### Implicazioni pratiche

- In fase iniziale, **è più efficace migliorare i dati che il modello**: pulizia, quantità, rappresentatività, e varietà sono spesso più importanti della scelta dell’algoritmo.
- È spesso vantaggioso iniziare con **modelli semplici**, come regressione lineare, decision tree o naive Bayes, per costruire un primo sistema funzionante.
- Solo quando i dati raggiungono un certo livello di saturazione o complessità, ha senso **investire in modelli più sofisticati** come reti neurali profonde o ensemble complessi.

---


````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dataset iniziale: paesi moderatamente ricchi
# x = reddito pro capite (in migliaia), y = livello di felicità (da 0 a 10)
X_base = np.array([[10], [15], [20], [25], [30], [35], [40]])
y_base = np.array([5.0, 5.5, 6.0, 6.5, 6.6, 6.4, 6.3])

# Nuovi paesi aggiunti: estremamente poveri o estremamente ricchi
X_extra = np.array([[2], [5], [70], [80]])
y_extra = np.array([6.2, 6.1, 5.8, 5.5])

# Combinazione dei due dataset
X_full = np.vstack((X_base, X_extra))
y_full = np.concatenate((y_base, y_extra))

# Modello 1: solo dati base
model_base = LinearRegression()
model_base.fit(X_base, y_base)
y_pred_base = model_base.predict(X_base)

# Modello 2: dati completi
model_full = LinearRegression()
model_full.fit(X_full, y_full)
x_range = np.linspace(0, 85, 100).reshape(-1, 1)
y_pred_full = model_full.predict(x_range)

# Grafico
plt.figure(figsize=(10, 6))
plt.scatter(X_base, y_base, color='blue', label='Dati base')
plt.scatter(X_extra, y_extra, color='green', label='Dati aggiunti')
plt.plot(X_base, y_pred_base, linestyle='dotted', color='blue', label='Modello base')
plt.plot(x_range, y_pred_full, linestyle='solid', color='red', label='Modello completo')
plt.xlabel('Reddito pro capite (in migliaia di $)')
plt.ylabel('Livello di felicità')
plt.title('Influenza di un campione più rappresentativo sul modello')
plt.legend()
plt.grid(True)
plt.show()
````

## Capitolo 4 – L’importanza della Rappresentatività dei Dati: Generalizzazione e Bias di Campionamento

Uno degli obiettivi fondamentali nel Machine Learning è la **generalizzazione**, ovvero la capacità di un modello di fare buone previsioni su dati mai visti prima. Affinché questo avvenga, **il dataset di addestramento deve essere rappresentativo dei casi reali su cui il modello verrà applicato**.

### Generalizzazione: cos’è e perché è importante

Generalizzare significa non solo memorizzare i dati di addestramento, ma imparare da essi in modo da fare previsioni corrette su nuovi esempi. Un modello che ha imparato troppo bene dai dati di partenza (overfitting) rischia di fallire su dati nuovi. Un altro rischio, meno evidente ma altrettanto critico, è **usare dati non rappresentativi**: in questo caso, il modello impara su una realtà che non corrisponde a quella in cui verrà applicato.

### Un esempio concreto

Supponiamo di voler prevedere il livello di felicità delle nazioni in funzione del reddito medio. Se il nostro dataset di addestramento contiene solo una selezione limitata di paesi — magari solo paesi europei — il modello risultante sarà influenzato da questa scelta.

Nel momento in cui si aggiungono paesi mancanti, si può osservare un cambiamento significativo nella forma del modello: ad esempio, potrebbe emergere che **i paesi molto ricchi non sono necessariamente più felici di quelli moderatamente ricchi**, e che **alcuni paesi poveri risultano sorprendentemente felici**. Questo comporta due conseguenze importanti:

1. **Il modello precedente non era affidabile**, perché basato su un campione non rappresentativo.
2. **Anche un modello semplice come una regressione lineare può diventare fuorviante** se i dati non riflettono la complessità del mondo reale.

### Cos’è un campione rappresentativo

Un dataset è rappresentativo quando **riflette fedelmente la varietà, la distribuzione e le caratteristiche della popolazione reale** che si vuole modellare. Questo include:

- Varietà nelle classi (es. ricchi, poveri, medi)
- Diversità geografica, culturale, temporale, ecc.
- Presenza di casi rari ma rilevanti

Se un campione è troppo piccolo, può generare **rumore statistico** (ossia fluttuazioni casuali che non riflettono la realtà). Ma anche **campioni molto grandi possono essere non rappresentativi** se il metodo di selezione è distorto. Questo problema si chiama **bias di campionamento**.

### Bias di campionamento

Il bias di campionamento si verifica quando il metodo con cui vengono raccolti i dati favorisce certi esempi a discapito di altri, rendendo il dataset sbilanciato o parziale. Alcuni esempi comuni:

- Raccogliere solo dati da utenti attivi (ignorando quelli silenziosi)
- Usare dati storici che riflettono decisioni passate distorte
- Ottenere feedback solo da chi risponde (bias di autoselezione)

Questi bias portano alla costruzione di modelli che **non generalizzano bene** e che spesso riflettono pregiudizi insiti nei dati originali.

### Conclusione

Per costruire modelli predittivi affidabili, **non basta avere molti dati: è essenziale che siano dati giusti**, cioè rappresentativi. Il processo di raccolta e selezione dei dati è quindi tanto importante quanto la scelta dell’algoritmo. Prima di addestrare un modello, occorre sempre porsi la domanda:

> “Questi dati riflettono realmente i casi su cui voglio applicare il modello?”

## Capitolo 5 – Bias di Campionamento: Esempi Reali e Implicazioni

Il **bias di campionamento** è uno dei pericoli più insidiosi nel Machine Learning. Si verifica quando il campione di dati utilizzato per l’addestramento non rappresenta correttamente la popolazione reale che si vuole modellare. Questo porta a modelli inaccurati e fuorvianti, anche quando i dati sono numerosi e il modello è tecnicamente ben costruito.

### Caso storico: Elezioni presidenziali USA del 1936

Uno degli esempi più famosi di bias di campionamento si è verificato durante le elezioni presidenziali americane del 1936, tra Alf Landon e Franklin D. Roosevelt. La rivista **Literary Digest** condusse un sondaggio su scala enorme:

- Furono inviate circa **10 milioni di lettere**.
- Risposero **2,4 milioni di persone**, un numero elevatissimo.
- Il sondaggio prevedeva che **Landon avrebbe vinto con il 57%** dei voti.

In realtà, Roosevelt vinse con **il 62%**, una differenza enorme.

#### Perché il modello ha sbagliato?

1. **Fonte dei dati distorta:** gli indirizzi per il sondaggio provenivano da elenchi telefonici, abbonamenti a riviste e club privati. Tutte fonti che tendevano a includere persone **più ricche** e quindi **più propense a votare per i Repubblicani**. I cittadini più poveri, che votavano in massa per Roosevelt, erano sottorappresentati.

2. **Nonresponse bias (bias di non risposta):** meno del 25% delle persone contattate rispose. Questo introdusse un’ulteriore distorsione: chi non rispose potrebbe essere stato **meno interessato alla politica**, **più povero**, **più scettico** o **non rappresentato nelle fonti**.

Questo caso dimostra che **un grande numero di risposte non garantisce accuratezza se il campione non è rappresentativo**. Il problema non era la dimensione del dataset, ma la **sua composizione**.

---

### Caso moderno: riconoscimento di video musicali

Immaginiamo ora di voler costruire un modello per riconoscere **video di musica funk** su YouTube. Un approccio semplice è cercare “funk music” e raccogliere i video ottenuti nei risultati. Ma anche qui si introduce un bias significativo:

- I risultati di ricerca di YouTube sono **ottimizzati per la popolarità**, non per la rappresentatività. Quindi il dataset sarà dominato da **artisti famosi**, ignorando magari artisti indipendenti o regionali.
- Inoltre, YouTube personalizza i risultati in base alla **posizione geografica**. Se l’utente si trova in Brasile, riceverà molti video di “**funk carioca**”, uno stile locale che è molto diverso dal funk classico americano (James Brown, Parliament-Funkadelic). Il modello potrebbe quindi **confondere due generi musicali completamente diversi**.

Questo esempio mostra come anche in contesti digitali moderni, **la fonte dei dati può introdurre bias nascosti**.

---

### Lezione chiave

La lezione che si trae da questi esempi è chiara:  
> **Non è sufficiente avere tanti dati. È necessario avere i dati giusti.**

Anche dataset molto grandi possono portare a conclusioni errate se sono **costruiti male**. Il bias di campionamento può essere difficile da individuare e spesso si nasconde nella selezione delle fonti, nei criteri di inclusione o nei comportamenti degli utenti.

---

### Tipi comuni di sampling bias

- **Selection bias:** il campione viene selezionato in modo non casuale.
- **Nonresponse bias:** solo una parte risponde, e chi risponde è sistematicamente diverso da chi non risponde.
- **Coverage bias:** alcuni gruppi della popolazione non hanno possibilità di essere inclusi nel campione.
- **Observer bias:** il ricercatore seleziona inconsapevolmente i dati che confermano la propria ipotesi.

## Capitolo 6 – Dati di Scarsa Qualità e Ingegneria delle Feature

Anche il miglior algoritmo non potrà mai produrre buoni risultati se viene alimentato con dati di bassa qualità. Un principio fondamentale del Machine Learning è:

> **Garbage in, garbage out**  
> Se i dati di input sono scadenti, anche l’output lo sarà.

### Dati di scarsa qualità: un ostacolo alla generalizzazione

Quando i dati di addestramento contengono errori, valori anomali o rumore (dovuto per esempio a misurazioni imprecise), il modello farà fatica a riconoscere i pattern reali. I dati rumorosi aumentano la variabilità e rendono più probabile l’overfitting.

Pulire i dati è una fase essenziale di qualsiasi progetto di Machine Learning. Non a caso, la maggior parte del tempo di un data scientist è spesso dedicata alla **preparazione e pulizia del dataset**.

### Esempi di dati problematici

#### Outlier

Un **outlier** è un’osservazione che si discosta fortemente dagli altri dati. Ad esempio, se stai analizzando i salari dei dipendenti e trovi un valore di 1 milione di euro, potrebbe trattarsi di un errore di inserimento.

**Soluzioni:**
- Rimuovere l’outlier se chiaramente errato
- Analizzarlo separatamente
- Correggere manualmente se conosci il valore corretto

#### Valori mancanti

È frequente trovare dati **mancanti** in uno o più campi. Ad esempio, se il 5% dei clienti non ha indicato l’età, occorre prendere una decisione:

**Possibili approcci:**
- Eliminare le istanze con valori mancanti
- Imputare valori (es. usare la **mediana** o **media** del campo)
- Costruire due modelli: uno con quella feature e uno senza
- Trattare i valori mancanti come una categoria separata

---

### Feature irrilevanti: il peso della qualità rispetto alla quantità

Un altro rischio è addestrare il modello su **feature irrilevanti**, cioè informazioni che non contribuiscono a spiegare il fenomeno studiato. Avere troppe feature inutili può aumentare la complessità del modello, causare overfitting e rallentare l’addestramento.

Per costruire un modello efficace serve **una buona rappresentazione dei dati**, ottenuta attraverso il processo chiamato **feature engineering**.

---

### Feature Engineering: costruire le giuste caratteristiche

Il feature engineering è l’arte di trasformare i dati grezzi in un set di variabili significative per l’algoritmo. Si articola in tre attività principali:

#### 1. Feature Selection

Selezionare le feature più utili tra quelle disponibili. Questo può essere fatto:
- Con metodi statistici (es. correlazione)
- Con test automatici (es. `SelectKBest`)
- Basandosi sulla conoscenza del dominio

#### 2. Feature Extraction

Combinare feature esistenti per crearne di nuove più informative. Ad esempio:
- Combinare “altezza” e “peso” in un indice di massa corporea (BMI)
- Usare tecniche di **riduzione dimensionale** come PCA per sintetizzare molte feature in poche componenti principali

#### 3. Creazione di nuove feature

Raccogliere dati aggiuntivi che migliorino il modello. A volte, **aggiungere la feature giusta è più utile che ottimizzare l’algoritmo**. Ad esempio:
- Integrare dati geografici (cap, latitudine, regione)
- Aggiungere dati di interazione (es. numero di acquisti per cliente)

---

### Conclusione

L’efficacia di un modello dipende in gran parte **dalla qualità dei dati e delle feature**, non solo dall’algoritmo. Investire tempo nella pulizia dei dati e nella progettazione delle giuste variabili è una delle attività più produttive in un progetto di Machine Learning.

## Capitolo 7 – Test, Validazione e Generalizzazione

Per sapere se un modello di Machine Learning sarà efficace nel mondo reale, non basta analizzare quanto bene funziona sui dati su cui è stato addestrato. È necessario **valutare la sua capacità di generalizzare**, cioè fare previsioni accurate su nuovi dati mai visti prima.

### Generalizzazione: come si misura

L’unico modo affidabile per sapere quanto bene un modello si comporterà su nuovi dati è **testarlo su dati che non ha mai visto**. Questo può essere fatto in due modi:

1. **In produzione:** mettere il modello online e vedere come si comporta nel tempo. Questo approccio fornisce risultati reali ma comporta rischi, specialmente se il modello è poco accurato: utenti insoddisfatti, errori critici e perdita di fiducia.

2. **Con dati separati:** suddividere il dataset iniziale in due o più parti, in modo da poter addestrare e testare il modello prima di metterlo in produzione.

---

### Il metodo standard: training set e test set

Il metodo più semplice e comune è dividere i dati in:

- **Training set:** i dati su cui il modello viene addestrato
- **Test set:** i dati completamente separati che servono a valutare le performance

Una suddivisione tipica prevede:
- **80%** dei dati per il training
- **20%** dei dati per il test

Tuttavia, questa proporzione può variare in base alla dimensione del dataset. Se si dispone di **10 milioni di esempi**, anche **l’1% riservato al test** (100.000 esempi) può essere più che sufficiente per stimare con precisione la qualità del modello.

---

### Errore di generalizzazione

Il principale valore che ci interessa è **l’errore sul test set**, detto anche **errore di generalizzazione** (o *out-of-sample error*). Questo valore indica **quanto bene il modello riesce a fare previsioni su dati nuovi**.

Confrontando due valori chiave:

- **Training error:** l’errore commesso sui dati di addestramento
- **Test error (generalization error):** l’errore commesso sui nuovi dati

Possiamo ottenere indicazioni importanti:

#### Caso 1 – Training error basso, Test error basso  
Il modello ha appreso correttamente i pattern e generalizza bene. Questo è l’obiettivo ideale.

#### Caso 2 – Training error basso, Test error alto  
Il modello ha imparato troppo bene i dati di training, anche il rumore. È un caso di **overfitting**: il modello è troppo complesso e si adatta ai dati in modo eccessivo.

#### Caso 3 – Training error alto, Test error alto  
Il modello è incapace di apprendere i pattern. È un caso di **underfitting**: il modello è troppo semplice o mal addestrato.

---

### Best Practice: Validazione e Cross-Validation

Per migliorare l’affidabilità della valutazione, è comune usare un **terzo insieme di dati** chiamato **validation set**, oppure applicare la tecnica della **cross-validation**, in particolare quando il dataset è piccolo.

- **Validation set:** serve per regolare i parametri del modello e prevenire l’overfitting prima del test finale.
- **K-Fold Cross-Validation:** il dataset viene diviso in *k* blocchi, e il modello viene testato *k* volte, ogni volta con un blocco diverso come test set. La media dei risultati fornisce una stima più stabile dell’errore.

---

### Conclusione

Testare e validare correttamente un modello è **fondamentale per assicurarsi che funzioni nel mondo reale**. Una buona separazione dei dati, l’uso corretto del test set e una valutazione accurata dell’errore di generalizzazione sono passi essenziali in ogni progetto di Machine Learning.

## Capitolo 8 – Modelli, Assunzioni e il Teorema "No Free Lunch"

Un **modello** in Machine Learning è una **semplificazione** della realtà. Serve a catturare i pattern presenti nei dati ignorando i dettagli irrilevanti o casuali. L’obiettivo è costruire una rappresentazione che generalizzi bene, ossia che funzioni anche su dati mai visti.

### Perché i modelli semplificano?

I dati del mondo reale sono complessi, rumorosi e pieni di variabili incontrollabili. Un buon modello **non cerca di replicare tutto**, ma di **concentrarsi sulle informazioni essenziali**, ignorando il rumore. Questo avviene sempre attraverso delle **assunzioni** implicite o esplicite.

#### Esempio: il modello lineare

Un modello lineare assume che **esista una relazione lineare tra input e output**, e che **la distanza dei punti dalla retta** (ossia gli errori) sia rumore casuale, quindi trascurabile.

In realtà, raramente i dati reali seguono una relazione perfettamente lineare. Tuttavia, la **linearità è una semplificazione utile** in molti casi, soprattutto quando si ha bisogno di un modello interpretabile e veloce da addestrare.

---

### Il Teorema "No Free Lunch"

Nel 1996, il ricercatore **David Wolpert** pubblicò un risultato fondamentale per l’intera teoria dell’apprendimento automatico: il **No Free Lunch Theorem** (Teorema del "nessun pranzo gratis").

#### Cosa afferma il teorema?

> **Se non si fanno assunzioni sui dati, non esiste alcun modello che funzioni meglio di altri in media su tutti i possibili problemi.**

Questo significa che:
- Non esiste un **modello universale** valido per tutti i dataset
- Ogni modello ha prestazioni ottimali **solo su determinati tipi di dati**
- **Non puoi sapere in anticipo quale sarà il modello migliore**, a meno che tu non conosca qualcosa sulla struttura dei dati

#### Implicazioni pratiche

- Per problemi semplici, è ragionevole provare modelli lineari, con o senza regolarizzazione
- Per problemi complessi (es. immagini, audio, linguaggio), conviene testare **reti neurali** o modelli più avanzati
- Non esiste un unico approccio valido per tutto: la **scelta del modello dipende dai dati, dal problema e dalle assunzioni che sei disposto a fare**

---

### Come si sceglie un modello nella pratica?

Poiché non è possibile provare tutti i modelli esistenti, si adottano strategie pratiche:

1. **Fare ipotesi ragionevoli sui dati** (es. linearità, indipendenza, dimensionalità)
2. **Scegliere una famiglia di modelli** coerente con queste ipotesi
3. **Valutare empiricamente le performance** con esperimenti controllati (train/test/cross-validation)
4. **Iterare**: se le prestazioni non sono buone, si cambiano le assunzioni o si migliora la qualità dei dati

---

### Conclusione

Ogni modello semplifica la realtà sulla base di assunzioni. Alcune assunzioni funzionano bene in certi contesti, altre no. Il Teorema del No Free Lunch ci ricorda che **non esiste una soluzione universale**, ma che la scelta del modello va sempre guidata dalla conoscenza del problema e dai dati a disposizione.

## Capitolo 11 – Selezionare una Metrica di Valutazione: RMSE, MAE e le Norme

Una volta definito il problema e compreso il contesto, è fondamentale scegliere **una metrica per misurare le performance del modello**. Questa scelta ha un impatto diretto su:

- La valutazione della bontà del modello  
- Il confronto tra diversi modelli  
- Il tuning degli iperparametri  
- La comunicazione dei risultati al team e ai decisori

---

### Obiettivo del nostro modello

Nel caso del progetto che stiamo trattando (prevedere il prezzo medio delle abitazioni in un distretto), stiamo affrontando un **problema di regressione**. Quindi, la metrica ideale dovrà misurare **quanto le nostre previsioni numeriche si discostano dai valori reali**.

---

### RMSE – Root Mean Square Error

La metrica più comunemente usata per i problemi di regressione è l’**errore quadratico medio**, o **RMSE**. Misura la **radice della media dei quadrati degli errori di previsione**, dando più peso agli errori grandi.

#### Formula:
$$
\text{RMSE}(X, h) = \sqrt{\frac{1}{m} \sum_{i=1}^{m} \left( h(x^{(i)}) - y^{(i)} \right)^2}
$$

- \( m \) = numero di esempi nel dataset di valutazione  
- \( x^{(i)} \) = vettore delle feature del *i*-esimo esempio  
- \( y^{(i)} \) = valore reale da prevedere  
- \( h(x^{(i)}\) = \hat{y}^{(i)} \) = valore predetto dal modello  

> L’RMSE è sensibile agli **outlier**: un singolo errore molto grande può aumentare drasticamente il valore della metrica.

---

### MAE – Mean Absolute Error

Una metrica alternativa è l’**errore assoluto medio**, o **MAE**. In questo caso, invece di elevare al quadrato gli errori, si calcola semplicemente il **valore assoluto medio**.

#### Formula:
$$
\text{MAE}(X, h) = \frac{1}{m} \sum_{i=1}^{m} \left| h(x^{(i)}) - y^{(i)} \right|
$$

> Il MAE è **più robusto agli outlier**, poiché **non amplifica** gli errori molto grandi come fa l’RMSE.

---

### Differenza tra RMSE e MAE

| Aspetto                  | RMSE                             | MAE                             |
|--------------------------|----------------------------------|---------------------------------|
| Sensibilità agli outlier | Alta (penalizza fortemente)      | Bassa (penalizza linearmente)   |
| Interpretazione          | Simile alla deviazione standard  | Media degli errori assoluti     |
| Norma corrispondente     | \( \ell_2 \) (euclidea)          | \( \ell_1 \) (di Manhattan)     |

#### Esempio pratico:
- RMSE penalizza un errore di **+10** molto più di due errori di **+5**  
- MAE li tratta in modo equivalente

---

### Le norme nel contesto Machine Learning

Entrambe le metriche sono casi particolari di **norme**, che misurano la distanza tra vettori (in questo caso: il vettore delle previsioni e quello dei valori reali).

- **\( \ell_2 \)-norm (RMSE):**
  $$
  \|v\|_2 = \sqrt{\sum_{i=1}^{n} v_i^2}
  $$
  Focalizza l’attenzione sugli **errori grandi**

- **\( \ell_1 \)-norm (MAE):**
  $$
  \|v\|_1 = \sum_{i=1}^{n} |v_i|
  $$
  Tratta tutti gli errori con **peso uguale**

- **\( \ell_\infty \)-norm:**
  Considera solo il **massimo errore assoluto**

---

### Quando usare RMSE o MAE?

- **RMSE**: quando gli errori grandi sono particolarmente critici o costosi  
- **MAE**: quando si vuole una valutazione più equa degli errori  
- In pratica, è spesso utile **calcolare entrambe le metriche** per avere un quadro più completo delle performance del modello

---

### Conclusione

Scegliere la giusta metrica di valutazione è **fondamentale per guidare lo sviluppo del modello**.  
L’RMSE è più sensibile agli errori gravi e tende a essere usato come metrica di default nei problemi di regressione,  
ma il MAE può essere preferito in presenza di outlier.

## Capitolo 13 – Creazione di un Set di Test: Campionamento Casuale, Stratificato e Prevenzione del Bias

Uno degli errori più comuni e dannosi nei progetti di machine learning consiste nell’utilizzare i dati di test troppo tardi nel processo. L’esposizione precoce ai dati di test, anche involontaria, può portare a decisioni modellistiche influenzate dai risultati osservati, generando stime eccessivamente ottimistiche delle performance. Per questa ragione, la creazione anticipata di un set di test indipendente è una buona pratica imprescindibile.

---

### 1. L’importanza del test set

La creazione di un test set sin dalle prime fasi del progetto consente di:

- Ottenere una **valutazione imparziale** delle prestazioni finali del modello
- Prevenire il fenomeno noto come **data snooping bias**
- Stabilire un **benchmark oggettivo** per confrontare diversi approcci modellistici

---

### 2. Divisione casuale semplice

Il metodo più immediato per suddividere il dataset in training e test set è l’estrazione casuale di una porzione dei dati:

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

Questo approccio, sebbene semplice e spesso efficace, non garantisce che il test set sia rappresentativo di caratteristiche chiave del dataset, come il livello di reddito. Nei casi in cui alcune variabili siano fortemente sbilanciate o particolarmente rilevanti, questo può introdurre **bias di campionamento**.

---

### 3. Campionamento stratificato

Per mantenere la rappresentatività rispetto a una variabile cruciale, come il **reddito medio** (`median_income`), è preferibile adottare una **strategia di campionamento stratificato**. Si inizia categorizzando il reddito medio in fasce:

```python
import numpy as np
import pandas as pd

housing["income_cat"] = pd.cut(housing["MedInc"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
```

A questo punto è possibile generare training e test set preservando le proporzioni di ciascuna categoria:

```python
from sklearn.model_selection import train_test_split

strat_train_set, strat_test_set = train_test_split(
    housing,
    test_size=0.2,
    stratify=housing["income_cat"],
    random_state=42
)
```

Il vantaggio principale di questa strategia è che il test set rifletterà fedelmente la distribuzione del reddito osservata nel dataset completo.

---

### 4. Verifica della rappresentatività

È buona norma confrontare la distribuzione delle categorie nel test set stratificato con quella del dataset originale:

```python
print("Distribuzione nel dataset originale:")
print(housing["income_cat"].value_counts(normalize=True).sort_index())

print("\nDistribuzione nel test set (stratificato):")
print(strat_test_set["income_cat"].value_counts(normalize=True).sort_index())

print("\nDistribuzione nel test set (casuale):")
print(test_set["income_cat"].value_counts(normalize=True).sort_index())
```

Questa analisi confermerà l’efficacia dello stratified sampling nel preservare la proporzione di ciascun gruppo di reddito rispetto al campionamento casuale.

---

### 5. Rimozione della colonna ausiliaria

La colonna `income_cat` è stata introdotta unicamente a fini di stratificazione. Dopo l’uso, può essere rimossa:

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
```

---

### Conclusione

La suddivisione del dataset in un training set e un test set rappresentativo costituisce una fase fondamentale per garantire la validità delle valutazioni del modello. Il **campionamento stratificato** si rivela particolarmente utile quando è necessario conservare la distribuzione di variabili rilevanti. Adottare queste buone pratiche sin dalle prime fasi contribuisce alla costruzione di sistemi predittivi affidabili, robusti e generalizzabili.

---

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Ricrea la colonna per la stratificazione
housing["income_cat"] = pd.cut(housing["MedInc"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])

# Campionamento casuale
train_set_random, test_set_random = train_test_split(housing, test_size=0.2, random_state=42)

# Campionamento stratificato
strat_train_set, strat_test_set = train_test_split(
    housing,
    test_size=0.2,
    stratify=housing["income_cat"],
    random_state=42
)

# Calcola le proporzioni per ogni set
original_dist = housing["income_cat"].value_counts(normalize=True).sort_index()
random_dist = test_set_random["income_cat"].value_counts(normalize=True).sort_index()
strat_dist = strat_test_set["income_cat"].value_counts(normalize=True).sort_index()

# Unisci i dati in un DataFrame per visualizzazione
comparison_df = pd.DataFrame({
    "Originale": original_dist,
    "Stratificato": strat_dist,
    "Random": random_dist
})

# Grafico a barre affiancate
comparison_df.plot(kind="bar", figsize=(10, 6))
plt.title("Distribuzione delle categorie di reddito\nOriginale vs Campionamento Casuale vs Stratificato")
plt.xlabel("Fascia di reddito (income_cat)")
plt.ylabel("Proporzione")
plt.grid(True)
plt.xticks(rotation=0)
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
````

## Capitolo 17 – **Feature Scaling e Trasformazioni dei Dati**

**(Dataset: California Housing)**

Una volta gestite eventuali feature categoriche (reali o simulate), il passo successivo nella preparazione del dataset per il Machine Learning è **scalare e trasformare correttamente le variabili numeriche**. Questo passaggio è cruciale per garantire che gli algoritmi apprendano in modo efficace e bilanciato.

---

### 1. Perché scalare le feature?

Molti algoritmi di apprendimento automatico (come regressione lineare, SVM, reti neurali) sono **sensibili alla scala dei valori numerici**:

Nel dataset `fetch_california_housing`, ad esempio:

* `MedInc` (reddito mediano) varia da circa **0 a 15**
* `AveRooms` può superare **30.000**

Senza scaling, i modelli tendono a concentrarsi sulle feature con **valori numerici maggiori**, anche se meno rilevanti.

---

### 2. Due strategie comuni

#### 2.1 **Min-Max Scaling** (o normalizzazione)

Riporta ogni valore in un intervallo definito, ad esempio `[0, 1]`.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
housing_minmax = scaler.fit_transform(housing)
```

Questa tecnica è utile soprattutto per **reti neurali** o algoritmi che richiedono feature comprese in un intervallo definito.

---

#### 2.2 **Standardizzazione** (Z-score)

Porta ogni feature ad avere **media 0** e **deviazione standard 1**:

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
housing_std = std_scaler.fit_transform(housing)
```

È la tecnica più comune e robusta, soprattutto in presenza di **outlier**.

⚠️ **Nota**: `fit()` e `fit_transform()` devono essere applicati **solo al training set**, per evitare data leakage.

---

### 3. Trasformazione di feature sbilanciate

Alcune colonne, come `Population`, hanno una distribuzione **sbilanciata a destra** (right-skewed), cioè molti valori piccoli e pochi valori molto grandi.

#### Soluzioni comuni:

* **Logaritmo** per dati fortemente sbilanciati
* **Radice quadrata** per sbilanciamenti più leggeri
* **Bucketizzazione** (discretizzazione in classi)

```python
import numpy as np

housing["log_Population"] = np.log1p(housing["Population"])  # log(1 + x)
```

Visualizziamo la differenza:

```python
import matplotlib.pyplot as plt

housing["Population"].hist(bins=50, figsize=(8,4))
plt.title("Distribuzione originale")
plt.show()

housing["log_Population"].hist(bins=50, figsize=(8,4))
plt.title("Distribuzione log-scalata")
plt.show()
```

---

### 4. Trasformazioni complesse: RBF (Radial Basis Function)

Se una feature ha un **picco di interesse specifico**, possiamo creare una nuova colonna che misura la **somiglianza rispetto a un valore target**, usando il **kernel RBF**.

Esempio con `HouseAge` (età media delle abitazioni), supponendo un picco di interesse a 35:

```python
from sklearn.metrics.pairwise import rbf_kernel

age_simil_35 = rbf_kernel(housing[["HouseAge"]], [[35]], gamma=0.1)
```

Il risultato è una nuova feature che vale \~1 se `HouseAge ≈ 35`, e decresce rapidamente man mano che ci si allontana da 35.

---

### 5. Scalare anche il **target** (opzionale ma utile)

Se `MedHouseVal` (valore mediano delle abitazioni) ha una **distribuzione non gaussiana**, si può **scalare anche il target**, mantenendo la coerenza tra training e predizione.

```python
from sklearn.linear_model import LinearRegression
from sklearn.compose import TransformedTargetRegressor
from sklearn.preprocessing import StandardScaler

target_scaler = StandardScaler()
model = TransformedTargetRegressor(
    regressor=LinearRegression(),
    transformer=target_scaler
)

model.fit(housing[["MedInc"]], housing_labels)
predictions = model.predict(housing[["MedInc"]].iloc[:5])
```

Il modello lavora sul **target standardizzato** ma restituisce la **predizione nella scala originale**.

---

### 6. Bucketizzazione (discretizzazione)

In alcune situazioni, può essere utile **trasformare una variabile continua in fasce** numeriche o categoriche.

Ad esempio, possiamo suddividere `HouseAge` in 4 intervalli e assegnare un’etichetta numerica:

```python
housing["age_cat"] = pd.cut(housing["HouseAge"], bins=[0, 20, 35, 50, 100], labels=[1,2,3,4])
```

Questo approccio può migliorare la **robustezza** del modello in presenza di **comportamenti diversi tra fasce** (es. case nuove vs case molto vecchie).

---

## Conclusione

Il **Feature Scaling** e le **trasformazioni numeriche** sono elementi **fondamentali per la qualità del modello**:

* Riducono l'influenza indebita delle feature con valori alti
* Rendono i gradienti stabili nei modelli ottimizzati
* Migliorano la convergenza degli algoritmi
* Riducono l’impatto degli outlier
* Consentono di **adattare le distribuzioni** alle ipotesi dei modelli lineari o probabilistici

Questa fase si integra naturalmente nella **pipeline di preprocessing** e prepara il dataset per qualsiasi algoritmo, dai più semplici ai più avanzati.

## Capitolo 17.1 – Trasformazioni Complesse: RBF e Funzioni Non Lineari

Dopo aver applicato tecniche di *feature scaling* e trasformazioni logaritmiche per stabilizzare la distribuzione dei dati, possiamo considerare **trasformazioni più sofisticate** per modellare relazioni non lineari e complesse tra le variabili. Questo è particolarmente utile nei modelli lineari, dove la qualità dell'ingresso determina in larga misura la capacità predittiva del modello.

In questo capitolo ci concentriamo su una trasformazione particolarmente utile, la **Radial Basis Function (RBF)**, e introduciamo altre trasformazioni non lineari che possono essere implementate manualmente per arricchire il dataset.

---

### 1. Trasformazione RBF (Radial Basis Function)

La Radial Basis Function è una funzione gaussiana centrata su un punto specifico, che misura la "somiglianza" tra un valore numerico e un punto di riferimento. È definita dalla seguente formula:

$$
\text{RBF}(x, c) = \exp(-\gamma (x - c)^2)
$$

Dove:

* $x$ è il valore della feature,
* $c$ è il **centro** (il punto rispetto al quale si misura la distanza),
* $\gamma$ è un **iperparametro** che controlla la larghezza del "picco" (valori alti = picco più stretto).

#### Esempio: Similarità rispetto a `HouseAge = 35`

```python
from sklearn.metrics.pairwise import rbf_kernel

gamma = 0.1
housing["HouseAge_rbf_35"] = rbf_kernel(housing[["HouseAge"]], [[35]], gamma=gamma)
```

#### Visualizzazione:

```python
plt.figure(figsize=(8, 4))
plt.plot(housing["HouseAge"], housing["HouseAge_rbf_35"], '.', alpha=0.3)
plt.title("RBF Similarity rispetto a HouseAge = 35")
plt.xlabel("HouseAge")
plt.ylabel("Somiglianza (RBF)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Questa trasformazione è utile quando **una certa fascia di valori ha una rilevanza particolare**, ma la relazione non è lineare.

---

### 2. RBF multipli (Feature Expansion RBF)

È possibile **generare più trasformazioni RBF**, ciascuna centrata su un punto diverso, per creare una **mappa non lineare** dello spazio:

```python
centri = [[20], [35], [50]]
gamma = 0.1
for centro in centri:
    label = f"HouseAge_rbf_{centro[0]}"
    housing[label] = rbf_kernel(housing[["HouseAge"]], centro, gamma=gamma)
```

Questo approccio genera **più feature**, ciascuna attiva solo in una determinata regione del dominio. È una base utile per modelli lineari, polinomiali o SVM.

---

### 3. Altre trasformazioni complesse implementabili manualmente

#### 3.1. Funzioni sinusoidali

Le trasformazioni sinusoidali sono particolarmente utili per catturare **ciclicità o stagionalità**, come:

* Prezzi mensili (effetti stagionali)
* Distanza angolare
* Fattori orari nel tempo

```python
housing["sin_HouseAge"] = np.sin(housing["HouseAge"] / 10)
housing["cos_HouseAge"] = np.cos(housing["HouseAge"] / 10)
```

#### Visualizzazione:

```python
plt.figure(figsize=(10, 4))
plt.plot(housing["HouseAge"], housing["sin_HouseAge"], '.', alpha=0.3, label="sin")
plt.plot(housing["HouseAge"], housing["cos_HouseAge"], '.', alpha=0.3, label="cos")
plt.title("Trasformazioni Sinusoidali su HouseAge")
plt.xlabel("HouseAge")
plt.ylabel("Output")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

---

#### 3.2. Funzioni polinomiali e radici

Oltre al logaritmo, è utile applicare potenze frazionarie o quadrati per rendere simmetrica o normalizzare una feature:

```python
housing["sqrt_rooms"] = np.sqrt(housing["AveRooms"])
housing["rooms_squared"] = housing["AveRooms"] ** 2
```

#### 3.3. Soglie binarie (cutoff transformation)

Si possono creare feature **binare** che segnalano il superamento di una soglia critica:

```python
housing["high_income"] = (housing["MedInc"] > 6).astype(int)
```

Questa trasformazione è utile quando si sospetta che un **cambiamento qualitativo** avvenga oltre una certa soglia.

---

### 4. Quando usare trasformazioni complesse?

Queste trasformazioni sono più utili quando:

* Si utilizzano **modelli lineari** (Regressione Lineare, Lasso, Ridge)
* I dati **non seguono distribuzioni gaussiane**
* Si sospettano **pattern localizzati** o **ciclici**
* Si desidera migliorare la separabilità o la capacità del modello senza modificarne la natura

Tuttavia, devono essere applicate con cautela: **aumentano la dimensionalità** e possono introdurre **rumore o collinearità**.

---

## Conclusione

Le trasformazioni complesse permettono di catturare relazioni non lineari nei dati, **aumentando la capacità espressiva del dataset**. L'uso di funzioni come RBF, seno, logaritmo o soglie consente di modellare fenomeni altrimenti non rilevabili con modelli standard.

In un contesto professionale, **l’intuizione del dominio** e l’analisi esplorativa guidano la scelta delle trasformazioni da applicare. In fase di produzione, è fondamentale integrare queste trasformazioni in una **pipeline** per garantire coerenza e riproducibilità.

### 4. Costruire un trasformatore personalizzato con classe Python

Quando una trasformazione richiede **un’operazione di “addestramento”**, ad esempio il calcolo della media o la creazione di cluster, è necessario definire una classe.

#### Esempio: clone di `StandardScaler`

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_array, check_is_fitted

class StandardScalerClone(BaseEstimator, TransformerMixin):
    def __init__(self, with_mean=True):
        self.with_mean = with_mean

    def fit(self, X, y=None):
        X = check_array(X)
        self.mean_ = X.mean(axis=0)
        self.scale_ = X.std(axis=0)
        self.n_features_in_ = X.shape[1]
        return self

    def transform(self, X):
        check_is_fitted(self)
        X = check_array(X)
        if self.with_mean:
            X = X - self.mean_
        return X / self.scale_
```

Questa classe può essere integrata in pipeline ed è compatibile con `GridSearchCV`.

### 5. Trasformazioni basate su clustering: `ClusterSimilarity`

Possiamo usare un modello come `KMeans` per identificare cluster geografici, e poi calcolare la similarità di ciascun distretto con i centroidi trovati.

```python
from sklearn.cluster import KMeans

class ClusterSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=10, gamma=1.0, random_state=None):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state

    def fit(self, X, y=None, sample_weight=None):
        self.kmeans_ = KMeans(n_clusters=self.n_clusters, random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self

    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)

    def get_feature_names_out(self, names=None):
        return [f"Cluster {i} similarity" for i in range(self.n_clusters)]
```

#### Uso con il nostro dataset:

```python
cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1.0, random_state=42)

                                           sample_weight=housing_labels)
```

Il risultato sarà una matrice con 10 colonne (una per cluster), e ogni riga indicherà la **similarità geografica del distretto** con ciascun centroide.

## Capitolo 18 – Pipeline di Trasformazione dei Dati

Nel processo di preparazione dei dati per l'apprendimento automatico, uno dei problemi principali è l’esecuzione ordinata e coerente delle trasformazioni. Ogni trasformazione deve essere eseguita nella giusta sequenza, ed è fondamentale evitare errori manuali. Per questo, **Scikit-Learn** mette a disposizione uno strumento potente: **`Pipeline`**.

### 1. Cos'è una Pipeline

Una **pipeline** è una sequenza ordinata di trasformazioni (e opzionalmente anche di modelli predittivi) che vengono applicate automaticamente e in modo trasparente ai dati.

Esempio base per colonne numeriche:
```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ("impute", SimpleImputer(strategy="median")),
    ("standardize", StandardScaler()),
])
```

Ogni step è identificato da un nome univoco e da un trasformatore. Il metodo `fit()` della pipeline applica tutti i `fit_transform()` in sequenza, fino all'ultimo elemento, su cui viene chiamato solo `fit()` se è un predittore.

### 2. Alternativa: `make_pipeline`

Per scrivere meno codice puoi usare `make_pipeline`, che assegna i nomi automaticamente:
```python
from sklearn.pipeline import make_pipeline

num_pipeline = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())
```

### 3. Uso della Pipeline

Dopo aver costruito la pipeline, puoi usarla direttamente per trasformare i dati:
```python
housing_num_prepared = num_pipeline.fit_transform(housing_num)
```

Se vuoi ottenere un `DataFrame` leggibile:
```python
import pandas as pd

df_housing_num_prepared = pd.DataFrame(
    housing_num_prepared,
    columns=num_pipeline.get_feature_names_out(),
    index=housing_num.index
)
```

### 4. Pipeline Miste con `ColumnTransformer`

Finora abbiamo gestito separatamente variabili numeriche e categoriche. Ma nella pratica è utile costruire una **unica pipeline** in grado di gestire entrambe le tipologie di colonne.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore")
)

preprocessing = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs),
])
```

Per selezionare le colonne in automatico puoi usare:
```python
from sklearn.compose import make_column_selector, make_column_transformer

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
    (cat_pipeline, make_column_selector(dtype_include=object)),
)
```

### 5. Trasformazioni Complesse con Pipeline Composita

Nel nostro caso, vogliamo costruire una pipeline che:
- Imputi i valori mancanti.
- Standardizzi le feature numeriche.
- One-hot encoder per le categoriche.
- Crei nuove variabili derivate (feature engineering).
- Applichi trasformazioni logaritmiche.
- Usi un kernel RBF per rilevare somiglianze geografiche.
- Restituisca una matrice NumPy finale pronta per i modelli.

Ecco il codice:

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics.pairwise import rbf_kernel

# Funzione per calcolo dei rapporti
def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(_, __):
    return ["ratio"]

def ratio_pipeline():
    return make_pipeline(
        SimpleImputer(strategy="median"),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler()
    )

# Trasformazione logaritmica
log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler()
)

# Cluster geografici
class ClusterSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=10, gamma=1.0, random_state=None):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state

    def fit(self, X, y=None, sample_weight=None):
        from sklearn.cluster import KMeans
        self.kmeans_ = KMeans(self.n_clusters, random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self

    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)

    def get_feature_names_out(self, names=None):
        return [f"Cluster {i} similarity" for i in range(self.n_clusters)]

# Pipeline di default per le altre numeriche
default_num_pipeline = make_pipeline(SimpleImputer(strategy="median"), StandardScaler())

# Preprocessing finale
preprocessing = ColumnTransformer([
    ("bedrooms", ratio_pipeline(), ["total_bedrooms", "total_rooms"]),
    ("rooms_per_house", ratio_pipeline(), ["total_rooms", "households"]),
    ("people_per_house", ratio_pipeline(), ["population", "households"]),
    ("log", log_pipeline, ["total_bedrooms", "total_rooms", "population", "households", "median_income"]),
    ("geo", ClusterSimilarity(n_clusters=10, gamma=1., random_state=42), ["latitude", "longitude"]),
    ("cat", cat_pipeline, make_column_selector(dtype_include=object)),
], remainder=default_num_pipeline)  # housing_median_age

housing_prepared = preprocessing.fit_transform(housing)
```

Ora hai una matrice NumPy pronta, con **24 feature trasformate**, già scalate e ingegnerizzate.

```python
preprocessing.get_feature_names_out()
```

### Conclusione

Le pipeline sono uno strumento essenziale per:

- Organizzare le trasformazioni in modo modulare e riutilizzabile.
- Evitare errori nella preparazione dei dati.
- Integrare facilmente nuovi step di preprocessing o feature engineering.
- Automatizzare la preparazione sia in fase di training che in fase di predizione.

````python
# complete_analysis_and_model_training.py

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, FunctionTransformer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error, r2_score
from pandas.plotting import scatter_matrix
from sklearn.metrics.pairwise import rbf_kernel

# 1. Caricamento e preparazione dati
raw = fetch_california_housing(as_frame=True)
housing = raw.frame.copy()
housing.columns = [col.capitalize() for col in housing.columns]  # Capitalizza i nomi

# 2. Campionamento stratificato per ‘MedInc’
housing["IncomeCat"] = pd.cut(housing["Medinc"],
                              bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                              labels=[1, 2, 3, 4, 5])
train_set, test_set = train_test_split(
    housing, test_size=0.2, random_state=42,
    stratify=housing["IncomeCat"]
)
for dataset in (train_set, test_set):
    dataset.drop("IncomeCat", axis=1, inplace=True)

# 3. Esplorazione dati (EDA) sul training set
print("Prime 5 righe:")
print(train_set.head(), "\n")

print("Informazioni sul dataset:")
print(train_set.info(), "\n")

print("Statistiche descrittive:")
print(train_set.describe(), "\n")

# Figure: istogrammi
train_set.hist(bins=50, figsize=(12, 8))
plt.tight_layout()
plt.show()

# Scatter geografico
train_set.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.2, grid=True)
plt.title("Distribuzione geografica (alpha=0.2)")
plt.show()

# Prezzi vs posizione e densità
train_set.plot(kind="scatter", x="Longitude", y="Latitude",
               s=train_set["Population"] / 100, 
               c="Medhouseval", cmap="jet",
               colorbar=True, alpha=0.4, figsize=(10, 7), grid=True)
plt.title("Prezzi e densità")
plt.show()

# Correlazioni
corr_matrix = train_set.corr()
print("Correlazione con Medhouseval:")
print(corr_matrix["Medhouseval"].sort_values(ascending=False), "\n")

# Scatter matrix
attrs = ["Medhouseval", "Medinc", "Averooms", "Houseage"]
scatter_matrix(train_set[attrs], figsize=(12, 8))
plt.tight_layout()
plt.show()

# MedInc vs MedHouseVal
train_set.plot(kind="scatter", x="Medinc", y="Medhouseval",
               alpha=0.1, grid=True)
plt.title("MedInc vs MedHouseVal")
plt.show()

# 4. Feature engineering
train_set["rooms_per_house"] = train_set["Averooms"] / train_set["Aveoccup"]
train_set["bedrooms_ratio"] = train_set["Avebedrms"] / train_set["Averooms"]

# 5. Separazione features/target
housing_labels = train_set["Medhouseval"].copy()
housing_features = train_set.drop("Medhouseval", axis=1)

# 6. Pipeline di preprocessing
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("std_scaler", StandardScaler()),
])
housing_prepared = num_pipeline.fit_transform(housing_features)

# 7. Trasformazioni aggiuntive (esempio RBF su ‘Houseage’)
rbf_transformer = FunctionTransformer(
    rbf_kernel, kw_args=dict(Y=[[35.]], gamma=0.1))
houseage_rbf = rbf_transformer.transform(housing_features[["Houseage"]])

# 8. Addestramento e valutazione modelli
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42)
}

print("\nRisultati modelli (cross-validated):")
for name, model in models.items():
    # cross_val_predict for full training set performance estimate
    preds = cross_val_predict(model, housing_prepared, housing_labels, cv=10)
    rmse = root_mean_squared_error(housing_labels, preds)
    r2 = r2_score(housing_labels, preds)
    print(f"{name}: RMSE = {rmse:.2f}, R² = {r2:.4f}")

````

## Capitolo 20 – Fine-Tuning del Modello

Una volta individuati uno o più modelli promettenti, è fondamentale **ottimizzarne le prestazioni attraverso la selezione dei migliori iperparametri**. Questa fase è chiamata **fine-tuning**, ed è cruciale per ottenere il massimo dal proprio modello senza sovradattamento.

---

### 1. Grid Search

Uno dei metodi più robusti è la **ricerca esaustiva a griglia** (`GridSearchCV`), che valuta tutte le combinazioni possibili di iperparametri su una griglia definita dall’utente.

#### Esempio: tuning di `RandomForestRegressor` in una pipeline completa

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor

full_pipeline = Pipeline([
    ("preprocessing", preprocessing),  # Include scaling, log, cluster similarity, etc.
    ("random_forest", RandomForestRegressor(random_state=42)),
])

param_grid = [
    {
        "preprocessing__geo__n_clusters": [5, 8, 10],
        "random_forest__max_features": [4, 6, 8]
    },
    {
        "preprocessing__geo__n_clusters": [10, 15],
        "random_forest__max_features": [6, 8, 10]
    },
]

grid_search = GridSearchCV(full_pipeline, param_grid, cv=3,
                           scoring="neg_root_mean_squared_error",
                           verbose=2, n_jobs=-1)
grid_search.fit(housing, housing_labels)
```

#### Migliori iperparametri:

```python
print("Migliori iperparametri trovati:")
print(grid_search.best_params_)
```

#### Analisi risultati:

```python
import pandas as pd

cv_results = pd.DataFrame(grid_search.cv_results_)
cv_results["mean_test_rmse"] = -cv_results["mean_test_score"]
cv_results = cv_results.sort_values(by="mean_test_rmse")
print(cv_results[["params", "mean_test_rmse"]].head())
```

---

### 2. Randomized Search

Quando lo spazio degli iperparametri è troppo ampio, la **ricerca casuale** (`RandomizedSearchCV`) è spesso più efficiente.

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
    "preprocessing__geo__n_clusters": randint(3, 50),

rnd_search = RandomizedSearchCV(full_pipeline, param_distributions=param_distribs,

```

### 3. Metodi ad Efficienza Crescente

### 4. Metodi Ensemble

Un altro approccio di fine-tuning consiste nel **combinare più modelli**:

* I modelli ensemble, come le foreste casuali, uniscono modelli deboli per produrre un modello robusto.
* Si possono anche combinare manualmente più modelli diversi (es. media tra regressione lineare e foresta casuale).

```python
pred1 = grid_search.best_estimator_.predict(housing)
```

## Capitolo 22 – Randomized Search, Ensemble e Analisi dei Migliori Modelli

Una volta identificato un modello promettente, il passo successivo è l’**ottimizzazione fine** degli iperparametri per massimizzarne le prestazioni. In questo capitolo vedremo un approccio alternativo alla grid search, il **Randomized Search**, e introdurremo metodi avanzati per analizzare e migliorare ulteriormente il modello finale.

---

### 1. Randomized Search: un'alternativa scalabile

La **Grid Search** esplora tutte le possibili combinazioni di valori iperparametrici forniti. Questo approccio è efficace per spazi ristretti ma diventa proibitivo quando il numero di combinazioni cresce esponenzialmente. 

In alternativa, **RandomizedSearchCV** esplora un numero fisso di combinazioni, scegliendole **casualmente** da distribuzioni specificate per ogni iperparametro. Questo permette di:

- Esplorare valori continui o ampi intervalli.
- Evitare che iperparametri poco rilevanti allunghino inutilmente il processo.
- Lavorare con budget di tempo/computazione predefiniti.

#### Esempio: ottimizzazione del nostro modello Random Forest

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
    'preprocessing__geo__n_clusters': randint(3, 50),
    'random_forest__max_features': randint(2, 20)
}

rnd_search = RandomizedSearchCV(
    full_pipeline,
    param_distributions=param_distribs,
    n_iter=10,
    cv=3,
    scoring='neg_root_mean_squared_error',
    random_state=42
)

rnd_search.fit(housing, housing_labels)
```

Con questo approccio possiamo esplorare un'ampia gamma di combinazioni senza doverle definire tutte manualmente.

---

### 2. Altri metodi: HalvingGridSearchCV e HalvingRandomSearchCV

Scikit-Learn fornisce anche varianti efficienti:

- **HalvingRandomSearchCV** e **HalvingGridSearchCV** utilizzano strategie di **successiva eliminazione**: 
  - Si parte con molte combinazioni valutate con risorse limitate.
  - Le migliori proseguono alla fase successiva con risorse crescenti.
  - Si ottengono risultati competitivi con un consumo computazionale ridotto.

---

### 3. Metodi Ensemble

Un'altra strategia di miglioramento consiste nel **combinare più modelli**. Questo approccio è noto come **ensemble learning** e può spesso superare le prestazioni del miglior modello singolo.

- Il caso più noto è il **Random Forest**, che aggrega molti alberi decisionali.
- È possibile combinare anche modelli diversi: ad esempio, si può creare un ensemble tra Random Forest e un k-Nearest Neighbors, mediare le loro predizioni, e ottenere migliori risultati globali.
  
Nel prossimo capitolo vedremo vari approcci ensemble in dettaglio.

### 4. Analisi del modello finale

Dopo il tuning, possiamo analizzare il miglior modello ottenuto:

```python
final_model = rnd_search.best_estimator_
```

Visualizziamo ora l'importanza delle feature:

```python

    reverse=True
```

Questa analisi può suggerire azioni migliorative:

- Investire su feature altamente informative (es. median_income o bedrooms_ratio).

### 5. Analisi degli errori

Oltre alle metriche aggregate, è fondamentale analizzare **dove e perché il modello sbaglia**:

- Esamina le istanze con errore più alto.
- Verifica la presenza di **outlier**, errori nei dati o categorie sotto-rappresentate.
- Verifica che il modello funzioni **su tutti i gruppi**: rurale/urbano, ricco/povero, Nord/Sud, ecc.

Suddividere il validation set in **sottogruppi per categorie** può rivelare debolezze sistematiche:

- Se un modello ha prestazioni mediocri su una certa regione o fascia di reddito, è necessario intervenire.
- Il modello non dovrebbe essere rilasciato in produzione se danneggia gruppi specifici.

### Conclusioni

- **RandomizedSearchCV** è lo strumento ideale per l’ottimizzazione su spazi ampi.
- I **modelli ensemble** possono portare a prestazioni superiori.
- **Analizzare i modelli** e i loro **errori specifici** permette di rifinire e correggere in profondità.
- L’obiettivo è costruire un modello non solo preciso in media, ma **robusto, equo e stabile**.


````python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

# Modello base
xgb = XGBRegressor(random_state=42, objective="reg:squarederror", n_jobs=-1)

# Spazio degli iperparametri
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [3, 5],
    "learning_rate": [0.05, 0.1],
    "reg_alpha": [0, 0.1, 1],     # Lasso
    "reg_lambda": [1, 5, 10],     # Ridge
}

# Grid search con cross-validation 5-fold
grid_search = GridSearchCV(
    xgb, param_grid, scoring="neg_root_mean_squared_error", cv=5,
    verbose=2, n_jobs=-1
)

# Fit sul dataset preprocessato
grid_search.fit(housing_prepared, housing_labels)

# Risultati
print("Migliori iperparametri trovati:")
print(grid_search.best_params_)
print(f"RMSE migliore: {-grid_search.best_score_:.2f}")
````
